## Loading the JSON file

Load the data from the file `disease_data.json`.

```python
#Your code here 
import json
f = open('disease_data.json')
data = json.load(f)
```

## Convert to a DataFrame

Create a DataFrame from the JSON file. Be sure to retrive the column names for the dataframe. (Search within the 'meta' key of the master dictionary.) The DataFrame should include all 42 columns.

```python
#Your code here
import pandas as pd
df = pd.DataFrame(data['data'])
columns = []
for item in data['meta']['view']['columns']:
    columns.append(item['name'])
df.columns = columns
df
```

```python
#we want overall --  recopy code from above using stratification category 'overall'
columns = ['LocationDesc', 'DataSource', 'DataValueType', 'Topic', 'Question', 'YearStart', 'YearEnd', 'DataValue']
df1 = df[(df['Question'] == 'Current asthma prevalence among adults aged >= 18 years') 
        & (df['StratificationCategory1'] == 'Overall')][columns]
df1
```

```python
#we are just going to use crude prevalence
columns = ['LocationDesc', 'DataSource', 'DataValueType', 'Topic', 'Question', 'YearStart', 'YearEnd', 'DataValue']
df1 = df[(df['Question'] == 'Current asthma prevalence among adults aged >= 18 years') 
        & (df['StratificationCategory1'] == 'Overall')
        & (df['DataValueType'] == 'Crude Prevalence')][columns]
df1.head()
```

```python
#check for duplicates in the location column
df1['LocationDesc'].duplicated().sum()
```

```python
#no more duplicates so we can go ahead and start graphing
#check type of DataValue column
df1['DataValue'].dtype
```

```python
df1.head()
```

```python
df1['DataValue'].value_counts(normalize=True).cumsum().head()
```

```python
df1['DataValue'] = df1['DataValue'].astype(float)
```

```python
df1['DataValue'].dtype
df1['LocationDesc'].value_counts().head()
```

```python
df1 = df1.sort_values(by='DataValue')
```

```python
#now it is in a float we can create a bar plot
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('ggplot')


x = df1['LocationDesc']
y = df1['DataValue']

fig = plt.figure(figsize=(12,5))
sns.barplot(x, y,)
plt.title('Crude Asthma Prevalence Rates for Males 18 & Older')
plt.xlabel('State / Country')
plt.ylabel('Crude Asthma Prevalence Rate (%)')
plt.xticks(rotation=90)
plt.show()

```
